## 2. Norms

$L_1$ Norm ì˜ ì •ì˜ëŠ” ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤. 

$|v| = |x_0| + |x_1| + ... + |x_n|$

$L_2$ Norm ì˜ ì •ì˜ëŠ” ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤.

$||v|| = \sqrt {{x_0}^2 + {x_1}^2 + ... + {x_n}^2}$

ì´ë¥¼ ì°¸ì¡°í•˜ì—¬ ì£¼ì–´ì§„ ë²¡í„°ì— ëŒ€í•´ì„œ $L_2$ Norm ì˜ ê°’ê³¼ $L_1$ Norm ì˜ ê°’ì˜ ì°¨ì´(**ì ˆëŒ€ê°’**)ë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“œì„¸ìš”.



```python
def NormDif(v): # v = np.array()
    L1_norm = np.sum(np.abs(v))
    L2_norm = np.sqrt(np.sum(np.square(v)))
    return np.abs(L1_norm - L2_norm)
```

## 3. Errors



- 2ê°œì˜ ë²¡í„°ì™€, errorì˜ íƒ€ì…ì„ ì…ë ¥ë°›ì•„ ê·¸ ê²°ê³¼ê°’ì„ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

- ì—¬ê¸°ì„œ errorì˜ íƒ€ì…ì€ `MSE`, `MAE` 2ê°€ì§€ ì…ë‹ˆë‹¤.



```python
def error(x, y, type) :
    # Mean Absolute Error: ì‹¤ì œê°’ê³¼ ì˜ˆì¸¡ê°’ì˜ ì°¨ì´ë¥¼ ì ˆëŒ€ê°’ìœ¼ë¡œ ë°”ê¾¸ê³  í‰ê· í™”
    if type == 'MAE': 
        return np.sum(np.abs(x-y)) / len(x)
    # Mean Squared Error: ì‹¤ì œê°’ê³¼ ì˜ˆì¸¡ê°’ì˜ ì°¨ì´ë¥¼ ì œê³±í•´ í‰ê· í™”
    if type == 'MSE':
        return np.sum(np.square(x-y)) / len(x)

# í…ŒìŠ¤íŠ¸
x = np.array([1,2,3])
y = np.array([4,5,6])
print(error(x, y, 'MAE'))
print(error(x, y, 'MSE'))
```

# 4. Inverse

- ì£¼ì–´ì§„ matrixì— ëŒ€í•´ì„œ inverse matrixë¥¼ ê³„ì‚°í•˜ì—¬ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”. 
- ì´ ë•Œ inverse ê³„ì‚°ì´ ë¶ˆê°€ëŠ¥ í•œ ê²½ìš° `-1` ì„ ë°˜í™˜í•©ë‹ˆë‹¤.

```python
def myInverse(m) : # m = np.array()
    if m.ndim == 1:
        return -1
    else:
        if np.linalg.det(m) == 0:
            return -1
        return np.linalg.inv(m)
# í…ŒìŠ¤íŠ¸
a = np.array([[1., 2.], [3., 4.]])
print(myInverse(a))
b = np.array([[1., 2.], [3., 6.]])
print(myInverse(b))
print(myInverse(np.array([1,2,3,4])))
```
